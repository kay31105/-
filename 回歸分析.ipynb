```python
import pandas as pd
import seaborn as sns
from pandas_datareader import data
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import csv
```

```python
esg={}
with open('C:/Users/kay31/Desktop/final/ESG表.csv',newline='', encoding="BIG5", errors='ignore') as csvfile:
  esg = csv.DictReader(csvfile)
  esg_df = pd.DataFrame(esg)
for column in esg_df:
  esg_df[column] = pd.to_numeric(esg_df[column],errors='coerce')
esg_df
```

```python
del esg_df['short_name']
del esg_df['zh_name']
del esg_df['sector']
del esg_df['stk_no']
```

```python
esg_df=esg_df.dropna()
```

```python
esg_df.corr().style.background_gradient(cmap="Blues")
```

```python
from sklearn.linear_model import Lasso,LassoCV
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.linear_model import Ridge,RidgeCV
```

```python
#設定訓練測試集 7:3
x_data = esg_df[['esg_score','年化夏普比率','年化波動度','平均Treynor ratio','平均現金股利率'
                 ,'平均本益比','總資產']]
y_data = esg_df[['全年平均收益率']]
train_x, test_x, train_y, test_y = train_test_split(x_data, y_data, test_size=0.3, random_state=42)
```

```python
from sklearn import preprocessing
#建立StandardScaler物件
zscore = preprocessing.StandardScaler()
# 資料標準化
train_x = zscore.fit_transform(train_x)
test_x = zscore.fit_transform(test_x)
train_y = zscore.fit_transform(train_y)
test_y = zscore.fit_transform(test_y)
```

```python
#Multivariable Regreesion
```

```python
from sklearn import linear_model
```

```python
regr=linear_model.LinearRegression()
regr.fit(train_x, train_y)
```

```python
regr.intercept_   #截距
```

```python
regr.coef_        #係數
```

```python
regr.score(train_x,train_y)
```

```python
regr.score(test_x,test_y)
```

```python
#lasso
```

```python
clf_lasso=linear_model.Lasso(alpha=0.03636)
clf_lasso.fit(train_x, train_y)
```

```python
#找最佳alpha值
clf_lasso = LassoCV(alphas=[1e-3, 1e-2, 1e-1, 1]).fit(train_x, train_y)
clf_lasso.score(train_x, train_y)
```

```python
clf_lasso.coef_
```

```python
clf_lasso.intercept_
```

```python
clf_lasso.score(train_x, train_y)
```

```python
clf_lasso.score(test_x, test_y)
```

```python
#ridge
```

```python
clf_ridge=linear_model.Ridge(alpha=10)
clf_ridge.fit(train_x,train_y)
```

```python
#找最佳alpha值
clf_ridge = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1]).fit(train_x, train_y)
clf_ridge.score(train_x, train_y)
```
