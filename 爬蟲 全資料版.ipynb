```python
import requests
from bs4 import BeautifulSoup
headers = {
    'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
    }
pagedata =[]
for i in range(0, 89):
  url = "https://irplatform.tdcc.com.tw/ir/zh/marketing/ajaxEsgInvest?offset="+ str(i*20) +"&max=20"
  print(url)
  res = requests.get(url,headers=headers)
  res.encooding = "utf-8"
  print(res.text)
  soup=BeautifulSoup(res.text,'lxml')
  soup.select_one('#search-result-table')
  print(res.json())
  pagedata.append(res.json()['rows'])
```

```python
headers = ['short_name', 'stk_no', 'zh_name','sector', 'environment_score', 'scoial_score', 'esg_score','iva_rating',
           'absolute_rating','esg_rating','spScore', 'ranking','short_name_en','position','moodyScore']
```

```python
import csv
```

```python
def IsValidData(data, req_key_list):
    for req_key in req_key_list:
        if data.get(req_key) == '-':
            return False
    return True
# 只要req_key_list的欄位=="-"，判定為無效資料
# IsValidData(data,['environment_score']) 選擇必要的資料欄位
```

```python
valid_pagedata=[]
for page in pagedata:
    res_page=[]
    for data in page:
        if IsValidData(data, ['esg_score']):
            res_page.append(data)
    valid_pagedata.append(res_page)
```

```python
#刪除遺失值的資料集
with open('ac.csv','w',newline='') as f:
    writedCsv = csv.DictWriter(f,headers)
    writedCsv.writeheader()
    for page in valid_pagedata:
        writedCsv.writerows(page)
```

```python
#完整資料集 
with open('ab.csv','w',newline='') as f:
    writedCsv = csv.DictWriter(f,headers)
    writedCsv.writeheader()
    for page in pagedata:
        writedCsv.writerows(page)
```
